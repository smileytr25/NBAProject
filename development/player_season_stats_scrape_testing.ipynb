```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np 
import unicodedata
```

```python
def remove_accents(name):
    if name is None:
        return None

    normalized = unicodedata.normalize("NFKD", name)
    return "".join(
        char for char in normalized
        if not unicodedata.combining(char)
    )
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_1950_per_game.html"
```

```python
player_pg_stats = pd.read_html(url, attrs={"id" : "per_game_stats"})[0]
```

```python
player_pg_stats["Player"] = player_pg_stats["Player"].apply(remove_accents)
```

```python
player_pg_stats = player_pg_stats.drop(columns=["Awards", "Rk"])
```
