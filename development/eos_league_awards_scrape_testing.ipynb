```python
import pandas as pd 
import numpy as np 
import requests 
from bs4 import BeautifulSoup, Comment
from io import StringIO
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2025.html#all_all_awards"
```

```python
r = requests.get(url)

print(r.status_code == 200)
```

```python
soup = BeautifulSoup(r.content, "html.parser")
wrapper = soup.find("div", id="all_all_awards")

table_comment = wrapper.find(
    string=lambda text: isinstance(text, Comment)
    and 'id="all_awards"' in text
)

table_soup = BeautifulSoup(table_comment, "html.parser")
table = table_soup.find("table", id="all_awards")
```

```python
raw_df = pd.read_html(StringIO(str(table)))[0]
```

```python
raw_df
```

```python
raw_df["Alternate Award Name"] = raw_df["Award"].str.split("(").str[1].str.split(")").str[0]
raw_df["Alternate Award Name"] = raw_df["Alternate Award Name"].fillna("None")
```

```python
raw_df["Award"] = raw_df["Award"].str.split("(").str[0].str.strip()
```

```python
raw_df
```

```python
wrapper = soup.find("div", id="all_all-nba")
all_nba_comment = wrapper.find(
    string = lambda text: isinstance(text, Comment)
    and 'id="div_all-nba"' in text
)

all_nba_soup = BeautifulSoup(all_nba_comment, "html.parser")
all_nba_div = all_nba_soup.find("div", id="div_all-nba")
```

```python
first_team = all_nba_div.find("div", id="all-nba_1")
second_team = all_nba_div.find("div", id="all-nba_2")
third_team = all_nba_div.find("div", id="all-nba_3")
```

```python
data = {"player" : [], "accolade" : []}

for x in first_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("1st Team All-NBA")

for x in second_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("2nd Team All-NBA")

for x in third_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("3rd Team All-NBA")
```

```python
wrapper = soup.find("div", id="all_all-defensive")
all_nba_comment = wrapper.find(
    string = lambda text: isinstance(text, Comment)
    and 'id="div_all-defensive"' in text
)

all_nba_soup = BeautifulSoup(all_nba_comment, "html.parser")
all_nba_div = all_nba_soup.find("div", id="div_all-defensive")

first_team = all_nba_div.find("div", id="all-defensive_1")
second_team = all_nba_div.find("div", id="all-defensive_2")

for x in first_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("1st Team All-Defensive")

for x in second_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("2nd Team All-Defensive")
```

```python
wrapper = soup.find("div", id="all_all-rookie")
all_nba_comment = wrapper.find(
    string = lambda text: isinstance(text, Comment)
    and 'id="div_all-rookie"' in text
)

all_nba_soup = BeautifulSoup(all_nba_comment, "html.parser")
all_nba_div = all_nba_soup.find("div", id="div_all-rookie")

first_team = all_nba_div.find("div", id="all-rookie_1")
second_team = all_nba_div.find("div", id="all-rookie_2")

for x in first_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("1st Team All-Rookie")

for x in second_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("2nd Team All-Rookie")
```

```python
wrapper = soup.find("div", id="all_in-seasontournamentall-tournamentteam")
all_nba_comment = wrapper.find(
    string = lambda text: isinstance(text, Comment)
    and 'id="div_in-seasontournamentall-tournamentteam"' in text
)

all_nba_soup = BeautifulSoup(all_nba_comment, "html.parser")
all_nba_div = all_nba_soup.find("div", id="div_in-seasontournamentall-tournamentteam")

first_team = all_nba_div.find("div", id="in-seasontournamentall-tournamentteam_1")

for x in first_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("1st Team In-Season Tournament All-Tournament")
```

```python
pd.DataFrame(data)
```

```python
wrapper = soup.find("div", id="all_all_star_game_rosters")
all_nba_comment = wrapper.find(
    string = lambda text: isinstance(text, Comment)
    and 'id="div_all_star_game_rosters"' in text
)

all_nba_soup = BeautifulSoup(all_nba_comment, "html.parser")
all_nba_div = all_nba_soup.find("div", id="div_all_star_game_rosters")

first_team = all_nba_div.find("div", id="all_star_game_rosters_1")
second_team = all_nba_div.find("div", id="all_star_game_rosters_2")
third_team = all_nba_div.find("div", id="all_star_game_rosters_3")

for x in first_team.find_all("p") + second_team.find_all("p") + third_team.find_all("p"):
    data["player"].append(x.text)
    data["accolade"].append("NBA All-Star")
```

```python
awards = pd.DataFrame(data)
awards["player"] = awards.player.str.split("(").str[0].str.strip()
```
